```python
import pandas as pd
import numpy as np

# 1. 타겟 값 매핑 딕셔너리 정의
target_map_specific = {
    'Dropout': 0,
    'Enrolled': 1,
    'Graduate': 2
}

# 2. 'Target' 열에 map 함수를 적용하여 'Target_Encoded' 새 열 생성
ml_df['Target_Encoded'] = ml_df['Target'].map(target_map_specific)

# 3. 결과 확인
print("--- 'Target' 열 특정 숫자 값으로 인코딩 완료 ---")
print(f"ml_df에 'Target_Encoded' 열 추가 후 형태: {ml_df.shape}")

print("--- 인코딩된 값 확인 ---")
print(f'dropout => {target_map_specific["Dropout"]}')
print(f'enrolled: => {target_map_specific["Enrolled"]}')
print(f'graduate: => {target_map_specific["Graduate"]}')

print("\n'Target_Encoded' 열의 고유값 및 분포:")
print(ml_df['Target_Encoded'].value_counts().sort_index())

print("\n원본 'Target' 열과 인코딩된 'Target_Encoded' 열 비교 (첫 5행):")
print(ml_df[['Target', 'Target_Encoded']].head())

# 인코딩 후 결측치가 생겼는지 확인 (매핑되지 않은 값이 있다면 NaN이 생김)
print(f"\n'Target_Encoded' 열의 결측치 개수: {ml_df['Target_Encoded'].isnull().sum()}")
```

### 스케일링

- 원-핫 인코딩되지 않은 수치형 컬럼만 선택해 스케일링 진행
- 스케일러는 StandardScaler

## 09. Feature Engineering / 모델 학습

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 1. 스케일링 할 수치형 컬럼들 식별 (원-핫 인코딩되지 않은 컬럼들)
numeric_cols_for_scaling = ['Age at enrollment',
                            'Curricular units 1st sem (approved)',
                            'Curricular units 1st sem (grade)',
                            'Curricular units 2nd sem (approved)',
                            'Curricular units 2nd sem (grade)']

scaler = StandardScaler()

# 3. X_train의 수치형 컬럼들에 대해 fit_transform 수행
#    주의: fit_transform은 NumPy 배열을 반환하므로, 다시 DataFrame으로 만들어줘야 컬럼명이 유지됨.
#    원래 DataFrame의 인덱스를 사용해야 나중에 합치기 편함.
X_train_scaled_numeric = scaler.fit_transform(X_train[numeric_cols_for_scaling])
X_train_scaled_numeric_df = pd.DataFrame(X_train_scaled_numeric, columns=numeric_cols_for_scaling, index=X_train.index)

# 4. X_test의 수치형 컬럼들에 대해 transform 수행 (X_train에서 학습된 scaler 사용)
X_test_scaled_numeric = scaler.transform(X_test[numeric_cols_for_scaling])
X_test_scaled_numeric_df = pd.DataFrame(X_test_scaled_numeric, columns=numeric_cols_for_scaling, index=X_test.index)

# 5. 스케일링된 수치형 컬럼들을 원래 X_train, X_test에 업데이트
#    먼저 기존 수치형 컬럼들을 삭제하고, 스케일링된 데이터프레임을 합치는 것이 안전할 수 있음.
#    또는, 직접 값을 할당해도 됨. 여기서는 직접 할당.
X_train[numeric_cols_for_scaling] = X_train_scaled_numeric_df
X_test[numeric_cols_for_scaling] = X_test_scaled_numeric_df


# 6. 결과 확인
print("--- ((새로운 전략 - 6단계)): 수치형 특성 스케일링 완료 ---")
print(f"X_train (스케일링 후)의 형태: {X_train.shape}") # 형태는 변하지 않음 (3539, 137)
print(f"X_test (스케일링 후)의 형태: {X_test.shape}")   # 형태는 변하지 않음 (885, 137)

print("\nX_train의 스케일링된 수치형 컬럼들 첫 3행:")
print(X_train[numeric_cols_for_scaling].head(3))

print("\nX_test의 스케일링된 수치형 컬럼들 첫 3행:")
print(X_test[numeric_cols_for_scaling].head(3))

# 스케일링 후 평균과 표준편차 확인 (X_train 기준, 거의 0과 1에 가까워야 함)
print("\nX_train 스케일링된 수치형 컬럼들의 평균:")
print(X_train[numeric_cols_for_scaling].mean())
print("\nX_train 스케일링된 수치형 컬럼들의 표준편차:")
print(X_train[numeric_cols_for_scaling].std())
```

### 머신 러닝 모델 평가
- 사용할 모델: (총 9개)
  - "Logistic Regression"
  - "Decision Tree"
  - "Random Forest"
  - "Support Vector Machine"
  - "K-Nearest Neighbors"
  - "Gradient Boosting"
  - "XGBoost"
  - "LightGBM"
  - "CatBoost"

```python
# CatBoostClassifier 설치
# !pip install catboost
```

```python
import pandas as pd
import numpy as np

# 머신러닝 모델 라이브러리 임포트
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier # AdaBoost도 있었지!
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# 평가 지표 라이브러리 임포트
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. 모델 객체 생성 (기본 하이퍼파라미터 위주)
#    CatBoost는 verbose=0 옵션으로 학습 중 로그 출력을 줄일 수 있음.
models = {
    "Logistic Regression": LogisticRegression(solver='liblinear', multi_class='auto', random_state=42, max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42, n_jobs=-1), # n_jobs=-1로 병렬 처리 활용
    "Support Vector Machine": SVC(random_state=42, probability=True), # probability=True로 predict_proba 사용 가능하게 (시간 더 걸릴 수 있음)
    "K-Nearest Neighbors": KNeighborsClassifier(n_jobs=-1), # n_jobs=-1
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss', n_jobs=-1),
    "LightGBM": LGBMClassifier(random_state=42, n_jobs=-1, verbose=-1), # verbose=-1로 경고/로그 최소화
    "CatBoost": CatBoostClassifier(random_state=42, verbose=0) # verbose=0으로 학습 로그 끔
}

# 2. 각 모델 학습, 예측 및 평가
results = {} # 모델별 평가 결과를 저장할 딕셔너리

for model_name, model in models.items():
    print(f"--- 모델 학습 및 평가 시작: {model_name} ---")

    # 2a. 모델 학습 (X_train, y_train 사용)
    try:
        model.fit(X_train, y_train)
        print(f"{model_name} 학습 완료.")

        # 2b. 테스트 데이터로 예측 (X_test 사용)
        y_pred = model.predict(X_test)
        print(f"{model_name} 예측 완료.")

        # 2c. 성능 평가
        accuracy = accuracy_score(y_test, y_pred)
        # target_names는 LabelEncoder의 classes_ 순서와 동일하게
        le_classes_assumed = ['Dropout(0)', 'Enrolled(1)', 'Graduate(2)']
        class_report = classification_report(y_test, y_pred, target_names=le_classes_assumed, zero_division=0) # zero_division=0 추가
        conf_matrix = confusion_matrix(y_test, y_pred)

        results[model_name] = {
            "Accuracy": accuracy,
            "Classification Report": class_report,
            "Confusion Matrix": conf_matrix,
            "Model Object": model # 학습된 모델 객체도 저장
        }

        print(f"\n** {model_name} 평가 결과 **")
        print(f"정확도 (Accuracy): {accuracy:.4f}")
        print("\n분류 보고서 (Classification Report):\n", class_report)
        print("\n혼동 행렬 (Confusion Matrix):\n", conf_matrix)

    except Exception as e:
        print(f"{model_name} 처리 중 오류 발생: {e}")
        results[model_name] = {"Error": str(e)} # 오류 정보 저장

    finally:
        print("-" * 60 + "\n")


# 3. 모든 모델의 정확도 비교
print("\n--- 모든 모델 정확도 요약 ---")
accuracies_summary = {}
for model_name, result_data in results.items():
    if "Accuracy" in result_data:
        print(f"{model_name}: {result_data['Accuracy']:.4f}")
        accuracies_summary[model_name] = result_data['Accuracy']
    elif "Error" in result_data:
        print(f"{model_name}: 오류 발생 ({result_data['Error']})")

# 정확도 기준으로 정렬해서 보여주기
if accuracies_summary:
    print("\n--- 정확도 순 정렬 ---")
    sorted_accuracies = sorted(accuracies_summary.items(), key=lambda item: item[1], reverse=True)
    for model_name, acc in sorted_accuracies:
        print(f"{model_name}: {acc:.4f}")
```

### 상위 5개 모델 튜닝

- 상위 5개 모델 하이퍼파라미터 튜닝 (RandomizedSearchCV 사용)

```python
import pandas as pd
import numpy as np

# 모델 라이브러리
from catboost import CatBoostClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

# 하이퍼파라미터 튜닝 도구 및 평가 지표
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

# 1. 튜닝할 모델 및 파라미터 범위 정의
param_distributions = {
    "CatBoost": {
        'iterations': [100, 200, 300], # num_boost_round, n_estimators 와 유사
        'learning_rate': [0.01, 0.05, 0.1],
        'depth': [4, 6, 8], # max_depth
        'l2_leaf_reg': [1, 3, 5, 7, 9] # L2 정규화
    },
    "Support Vector Machine": {
        'C': [0.1, 1, 10, 100],
        'gamma': ['scale', 'auto', 0.01, 0.001], # 'rbf', 'poly', 'sigmoid' 커널에 유효
        'kernel': ['rbf', 'linear'] # 다양한 커널 시도
    },
    "LightGBM": {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'num_leaves': [20, 31, 40], # 중심 노드 최대 개수
        'max_depth': [-1, 5, 10], # -1은 제한 없음
        'subsample': [0.7, 0.8, 0.9], # 데이터 샘플링 비율 (bagging_fraction)
        'colsample_bytree': [0.7, 0.8, 0.9] # 특성 샘플링 비율 (feature_fraction)
    },
    "Gradient Boosting": {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5],
        'subsample': [0.7, 0.8, 0.9]
    },
    "XGBoost": {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5, 6],
        'subsample': [0.7, 0.8, 0.9],
        'colsample_bytree': [0.7, 0.8, 0.9],
        'gamma': [0, 0.1, 0.2] # 분할을 위한 최소 손실 감소
    }
}

# 튜닝 대상 모델 객체 (이름은 param_distributions의 key와 일치시켜야 함)
# 기본 random_state 등은 유지, n_jobs=-1 등도 추가 가능
base_models = {
    "CatBoost": CatBoostClassifier(random_state=42, verbose=0),
    "Support Vector Machine": SVC(random_state=42, probability=True), # probability=True 유지
    "LightGBM": LGBMClassifier(random_state=42, n_jobs=-1, verbose=-1),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss', n_jobs=-1)
}

# RandomizedSearchCV 설정
n_iter_search = 10  # 각 모델별로 시도할 파라미터 조합 수 (시간이 많으면 늘려도 좋음)
cv_folds = 3      # 교차 검증 폴드 수 (시간이 많으면 5 정도로 늘려도 좋음)

tuned_results = {}

# 2. for 루프를 사용하여 각 모델 튜닝
for model_name, params in param_distributions.items():
    print(f"--- 하이퍼파라미터 튜닝 시작: {model_name} ---")
    base_model = base_models[model_name]

    # RandomizedSearchCV 객체 생성
    random_search = RandomizedSearchCV(
        estimator=base_model,
        param_distributions=params,
        n_iter=n_iter_search,
        cv=cv_folds,
        scoring='accuracy', # 평가 지표 (정확도)
        random_state=42,
        n_jobs=-1, # 병렬 처리
        verbose=1  # 진행 상황 표시
    )

    # 튜닝 실행 (X_train, y_train 사용)
    try:
        random_search.fit(X_train, y_train)

        print(f"\n{model_name} 튜닝 완료.")
        print(f"최적 하이퍼파라미터: {random_search.best_params_}")
        print(f"최적 교차 검증 정확도: {random_search.best_score_:.4f}")

        # 최적 모델로 테스트 세트 예측 및 평가
        best_model = random_search.best_estimator_
        y_pred_tuned = best_model.predict(X_test)
        accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
        class_report_tuned = classification_report(y_test, y_pred_tuned,
                                                  target_names=['Dropout(0)', 'Enrolled(1)', 'Graduate(2)'],
                                                  zero_division=0)

        tuned_results[model_name] = {
            "Best Params": random_search.best_params_,
            "Best CV Score": random_search.best_score_,
            "Test Accuracy": accuracy_tuned,
            "Test Classification Report": class_report_tuned,
            "Best Estimator": best_model
        }
        print(f"테스트 세트 정확도 (튜닝 후): {accuracy_tuned:.4f}")
        print("\n테스트 세트 분류 보고서 (튜닝 후):\n", class_report_tuned)

    except Exception as e:
        print(f"{model_name} 튜닝 중 오류 발생: {e}")
        tuned_results[model_name] = {"Error": str(e)}
    finally:
        print("-" * 70 + "\n")


# 3. 모든 튜닝된 모델의 테스트 정확도 비교
print("\n--- 모든 튜닝된 모델 테스트 정확도 요약 ---")
tuned_accuracies_summary = {}
for model_name, result_data in tuned_results.items():
    if "Test Accuracy" in result_data:
        print(f"{model_name}: {result_data['Test Accuracy']:.4f} (CV Best: {result_data['Best CV Score']:.4f})")
        tuned_accuracies_summary[model_name] = result_data['Test Accuracy']
    elif "Error" in result_data:
        print(f"{model_name}: 오류 발생 ({result_data['Error']})")

# 정확도 기준으로 정렬해서 보여주기
if tuned_accuracies_summary:
    print("\n--- 튜닝 후 정확도 순 정렬 (Test Accuracy 기준) ---")
    sorted_tuned_accuracies = sorted(tuned_accuracies_summary.items(), key=lambda item: item[1], reverse=True)
    for model_name, acc in sorted_tuned_accuracies:
        print(f"{model_name}: {acc:.4f}")
```

- 정확도 0.9 이상 상위 5개 모델 하이퍼파라미터 튜닝 (RandomizedSearchCV 사용)

```python
import pandas as pd
import numpy as np

# 모델 라이브러리
from lightgbm import LGBMClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.svm import SVC

# 하이퍼파라미터 튜닝 도구 및 평가 지표
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

# 이전에 정의된 X_train_binary, X_test_binary, y_train_binary, y_test_binary를 사용.

# 1. 튜닝할 모델 및 파라미터 범위 정의
#    (이전 3클래스 튜닝 때의 범위를 참고하되, 필요시 약간 조절 가능)
param_distributions_binary = {
    "LightGBM": {
        'n_estimators': [100, 200, 300, 400],
        'learning_rate': [0.01, 0.05, 0.1],
        'num_leaves': [20, 31, 40, 50],
        'max_depth': [-1, 5, 10, 15],
        'subsample': [0.7, 0.8, 0.9, 1.0],
        'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
        'reg_alpha': [0, 0.01, 0.1], # L1 정규화
        'reg_lambda': [0, 0.01, 0.1] # L2 정규화
    },
    "Gradient Boosting": {
        'n_estimators': [100, 200, 300, 400],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5, 6],
        'subsample': [0.7, 0.8, 0.9, 1.0],
        'min_samples_split': [2, 4, 6],
        'min_samples_leaf': [1, 2, 3]
    },
    "XGBoost": {
        'n_estimators': [100, 200, 300, 400],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5, 6, 7],
        'subsample': [0.7, 0.8, 0.9, 1.0],
        'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
        'gamma': [0, 0.1, 0.2, 0.3],
        'reg_alpha': [0, 0.01, 0.1], # L1
        'reg_lambda': [1, 1.1, 1.2] # L2 (기본값 1에서 조금 변경)
    },
    "CatBoost": {
        'iterations': [100, 200, 300, 400],
        'learning_rate': [0.01, 0.05, 0.1],
        'depth': [4, 6, 8, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9],
        'border_count': [32, 64, 128] # 연속형 변수 분할 시 사용하는 그리드 수
    },
    "Support Vector Machine": {
        'C': [0.1, 1, 10, 50, 100], # C 값 범위 확장
        'gamma': ['scale', 'auto', 0.001, 0.0001], # gamma 값 범위 미세 조정
        'kernel': ['rbf', 'linear', 'poly'], # poly 커널 추가
        'degree': [2, 3, 4] # poly 커널 사용 시 차수 (kernel='poly'일 때만 의미 있음)
    }
}

# 튜닝 대상 모델 객체
base_models_binary_tune = {
    "LightGBM": LGBMClassifier(random_state=42, n_jobs=-1, verbose=-1),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', n_jobs=-1),
    "CatBoost": CatBoostClassifier(random_state=42, verbose=0),
    "Support Vector Machine": SVC(random_state=42, probability=True) # probability=True 유지
}

# RandomizedSearchCV 설정
n_iter_search_binary = 15  # 시도 횟수를 조금 늘려봄 (시간 고려하여 조절)
cv_folds_binary = 3        # 교차 검증 폴드 수

tuned_results_binary = {}

# 2. for 루프를 사용하여 각 모델 튜닝
for model_name in param_distributions_binary.keys(): # param_distributions_binary에 있는 모델만 튜닝
    if model_name not in base_models_binary_tune:
        print(f"경고: {model_name}에 대한 기본 모델 객체가 base_models_binary_tune에 없습니다. 건너<0xEB><0x9A><0xB0>니다.")
        continue

    print(f"--- 하이퍼파라미터 튜닝 시작 (이진 분류): {model_name} ---")
    base_model = base_models_binary_tune[model_name]
    params = param_distributions_binary[model_name]

    # RandomizedSearchCV 객체 생성
    # SVM의 경우 kernel='poly'일 때만 'degree'가 의미 있으므로, 이런 경우는 파이프라인이나 조건부 로직이 필요할 수 있음
    # 여기서는 일단 모든 조합을 시도하도록 단순화 (일부 조합은 경고 발생 가능)
    random_search_binary = RandomizedSearchCV(
        estimator=base_model,
        param_distributions=params,
        n_iter=n_iter_search_binary,
        cv=cv_folds_binary,
        scoring='accuracy',
        random_state=42,
        n_jobs=-1,
        verbose=1
    )

    try:
        random_search_binary.fit(X_train_binary, y_train_binary)

        print(f"\n{model_name} 튜닝 완료.")
        print(f"최적 하이퍼파라미터: {random_search_binary.best_params_}")
        print(f"최적 교차 검증 정확도: {random_search_binary.best_score_:.4f}")

        best_model_binary = random_search_binary.best_estimator_
        y_pred_tuned_binary = best_model_binary.predict(X_test_binary)
        accuracy_tuned_binary = accuracy_score(y_test_binary, y_pred_tuned_binary)
        class_report_tuned_binary = classification_report(y_test_binary, y_pred_tuned_binary,
                                                          target_names=['Dropout(0)', 'Graduate(1)'],
                                                          zero_division=0)

        tuned_results_binary[model_name] = {
            "Best Params": random_search_binary.best_params_,
            "Best CV Score": random_search_binary.best_score_,
            "Test Accuracy": accuracy_tuned_binary,
            "Test Classification Report": class_report_tuned_binary,
            "Best Estimator": best_model_binary
        }
        print(f"테스트 세트 정확도 (튜닝 후): {accuracy_tuned_binary:.4f}")
        # print("\n테스트 세트 분류 보고서 (튜닝 후):\n", class_report_tuned_binary) # 너무 길어질 수 있으니 일단 주석

    except Exception as e:
        print(f"{model_name} 튜닝 중 오류 발생: {e}")
        tuned_results_binary[model_name] = {"Error": str(e)}
    finally:
        print("-" * 70 + "\n")

# 3. 모든 튜닝된 모델의 테스트 정확도 비교
print("\n--- 모든 튜닝된 모델 테스트 정확도 요약 (이진 분류) ---")
tuned_accuracies_summary_binary = {}
for model_name, result_data in tuned_results_binary.items():
    if "Test Accuracy" in result_data:
        print(f"{model_name}: {result_data['Test Accuracy']:.4f} (CV Best: {result_data['Best CV Score']:.4f})")
        tuned_accuracies_summary_binary[model_name] = result_data['Test Accuracy']
    elif "Error" in result_data:
        print(f"{model_name}: 오류 발생 ({result_data['Error']})")

if tuned_accuracies_summary_binary:
    print("\n--- 튜닝 후 정확도 순 정렬 (Test Accuracy 기준) ---")
    sorted_tuned_accuracies_binary = sorted(tuned_accuracies_summary_binary.items(), key=lambda item: item[1], reverse=True)
    for model_name, acc in sorted_tuned_accuracies_binary:
        print(f"{model_name}: {acc:.4f}")
```

## 최종 모델 선택: LightGBM (Light Gradient Boosting Machine)

**선택 이유:**

*   **가장 높은 예측 정확도:** "Enrolled" 클래스를 제외하고 "Dropout"과 "Graduate"를 예측하는 이진 분류 문제로 전환한 후, 여러 머신러닝 모델을 비교 평가한 결과, LightGBM이 **테스트 데이터에서 약 91.18%의 가장 높은 정확도**를 보였습니다.

*   **균형 잡힌 성능:** 단순히 전체 정확도만 높은 것이 아니라, 분류 보고서(Classification Report)를 통해 확인했을 때 "Dropout" 클래스와 "Graduate" 클래스 모두에 대해 **높은 F1-Score (각각 0.88, 0.93)** 를 기록하여, 특정 클래스에 치우치지 않고 전반적으로 우수한 예측 성능을 나타냈습니다.

*   **효율성:** LightGBM은 대용량 데이터에서도 빠르고 효율적인 학습이 가능한 것으로 알려져 있으며, 이번 분석 과정에서도 다른 복잡한 모델들에 비해 비교적 빠른 학습 속도를 보여주었습니다.

**LightGBM의 간략한 원리:**

LightGBM은 **그래디언트 부스팅(Gradient Boosting)** 프레임워크를 기반으로 하는 트리 기반의 앙상블 학습 모델입니다. 핵심 아이디어는 다음과 같습니다:

1.  **여러 개의 결정 트리(Decision Tree) 사용:** 하나의 강력한 모델 대신, 예측 능력이 상대적으로 약한 여러 개의 결정 트리를 순차적으로 만들어 나갑니다.
2.  **오류 보완 학습:** 각 트리는 이전 트리가 만든 예측의 오류(잔차, residual)를 학습하여 보완하는 방향으로 생성됩니다. 즉, 앞선 모델이 잘 맞추지 못한 부분을 다음 모델이 더 집중적으로 학습합니다.
3.  **리프 중심 트리 분할 (Leaf-wise Tree Growth):** 다른 부스팅 모델들이 주로 사용하는 레벨 중심(level-wise) 트리 분할 방식과 달리, LightGBM은 손실(loss)을 가장 크게 줄일 수 있는 리프(leaf) 노드를 찾아 집중적으로 분할합니다. 이를 통해 더 적은 수의 리프로도 복잡한 패턴을 학습할 수 있어 효율적입니다.
4.  **경사 하강법 최적화:** 손실 함수를 최소화하기 위해 경사 하강법(Gradient Descent)을 사용하여 각 트리의 가중치와 구조를 최적화합니다.
5.  **효율성을 위한 기술:** GOSS(Gradient-based One-Side Sampling)와 EFB(Exclusive Feature Bundling) 같은 기술을 사용하여 학습 속도를 높이고 메모리 사용량을 줄입니다.

이러한 특징들 덕분에 LightGBM은 높은 예측 정확도와 함께 빠른 학습 속도를 제공하여, 다양한 분류 및 회귀 문제에서 널리 사용되는 강력한 알고리즘 중 하나입니다.
